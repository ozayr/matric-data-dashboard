```python
# pretty sure there is province info in center or emis num 
# sure enough there is a coding
province_by_centerNo = {4:"EASTERN CAPE",
3:"FREE STATE",
8:"GAUTENG",
5:"KWAZULU-NATAL",
7:"LIMPOPO",
6:"MPUMALANGA",
9:"NORTH WEST",
2:"NORTHERN CAPE",
1:"WESTERN CAPE"}

def make_province(center_no):
    return province_by_centerNo[int(str(center_no)[0])]

raw_data['Province'] = raw_data.centre_no.apply(make_province)
```
